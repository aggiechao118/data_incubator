```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
data = pd.DataFrame(pd.read_csv("shot_logs.csv"))
game_id = data["GAME_ID"].unique()
two = []
three = []
tot = []
for id in game_id:
    td = data[data["GAME_ID"]==id]
    two.append(td[td["PTS_TYPE"] == 2]["PTS"].sum());
    three.append(td[td["PTS_TYPE"] == 3]["PTS"].sum());
    tot.append(td["PTS"].sum())
plt.scatter(two, tot, label = "two-total",color='green', marker= "*", s=5) 
plt.xlabel('two_points')
plt.ylabel('total_socre')
plt.show()
```
